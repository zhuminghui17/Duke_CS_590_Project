```python
# Create aggregate summary for each iteration (combined later)
# Reading CSV files and attempt to create an average for each indicators on daily groupby
# Main variables to create average of:
# Trip seconds
# Trip miles
# Fare
# Tip
# Tip as percentage of (Fare + Additional Charge)
# Trips pooled
# Shared Trip Authorized

df_summary = pd.DataFrame()

for i in range(1, 301):
    
    # Reading the dataset one by one
    print("Reading the dataset number " + str(i))
    name = "D:\coding project\cs590\Trips" + str(i) + ".csv"
    df = pd.read_csv(name)
    
     #create "date" and "hours timestamp as base for mean"
    df['date'] = df['Trip Start Timestamp'].str[:10]    
    df['time_start'] = df['Trip Start Timestamp'].str[11:16]
    df['ampm'] = df['Trip Start Timestamp'].str[-2:]
    df['count'] = 1
    
    # Create tip as percentage of fare
    df['tipperc'] = df['Tip'] / (df['Fare'] + df['Additional Charges'])
    
    # Create the summary count of data
    df_summary = pd.concat([df_summary, df.groupby('date').agg({'Trip Seconds': 'mean', 'Fare': 'mean', 'Trip Miles': 'mean', 
                                                               'Tip': 'mean', 'tipperc': 'mean', 'Shared Trip Authorized': 'mean',
                                                               'count' : 'sum'})])
    
df_summary.to_csv("D:\coding project\cs590\TripsMeanStreaming.csv")
```

```python
import requests
import json
from datetime import datetime

import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
from pandas import Series, DataFrame

# 1. Make an API call and store the response.
def getWeatherInfo(lat, long, date, hour):
  url = f'https://archive-api.open-meteo.com/v1/archive?latitude={lat}' \
        +f'&longitude={long}&start_date={date}&end_date={date}&hourly=temperature_2m' \
        +'&hourly=temperature_2m,relativehumidity_2m,precipitation,rain,snowfall,' \
        +'windspeed_10m,winddirection_10m&timezone=America%2FChicago'
  try:
    data = requests.get(url)
    data = data.json()
    weatherInfo = {
    'tempHourly': data['hourly']['temperature_2m'][hour],
    'precipitationHourly': data['hourly']['precipitation'][hour],
    'rainHourly': data['hourly']['rain'][hour],
    'snowfallHourly': data['hourly']['snowfall'][hour],
    'winspeedHourly': data['hourly']['windspeed_10m'][hour],
    'windirectionHourly': data['hourly']['winddirection_10m'][hour]}
    return weatherInfo
  except:
    return data.json()
```

```python
import time
def addWeatherInfo(data):
  startTime = time.time()
  # generate empty columns
  for i in ['tempHourly', 'precipitationHourly', 'rainHourly', 'snowfallHourly', 'winspeedHourly', 'windirectionHourly']:
    data[i] = pd.Series()
  try:
    for i in range(len(data)):
      if (i+1) % 10000 == 0:
        currTime = time.time()
        print(f'Completed Obs: {(i*10000)}/3000000, Current Process: {round((i*10000)/3000000 * 100, 3)}%, Time Spent: {currTime-startTime}')
      info = getWeatherInfo(data['Pickup Centroid Latitude'][i], data['Pickup Centroid Longitude'][i], data['Date'][i], data['Hour'][i])
      for j in ['tempHourly', 'precipitationHourly', 'rainHourly', 'snowfallHourly', 'winspeedHourly', 'windirectionHourly']:
        data.at[i, j] = info[j]
  except:
    print(i)
  return data
```
